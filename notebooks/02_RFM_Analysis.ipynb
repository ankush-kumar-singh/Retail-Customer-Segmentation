## RFM SEGMENTATION

```python
# INdustry rfm segments
import numpy as np
```

```python
rfm_df[rfm_df["Segment"] == "Others"][
    ["Recency", "Frequency", "Monetary",
     "R_score", "F_score", "M_score", "Score"]
].head(30)
```

```python
rfm_df["R_score"] = rfm_df["R_score"].astype(int)
rfm_df["F_score"] = rfm_df["F_score"].astype(int)
rfm_df["M_score"] = rfm_df["M_score"].astype(int)
```

```python
conditions = [

    # Champions
    (rfm_df["R_score"] == 5) &
    (rfm_df["F_score"] >= 4) &
    (rfm_df["M_score"] >= 4),

    # Loyal Customers
    (rfm_df["R_score"] >= 4) &
    (rfm_df["F_score"] >= 4),

    # Potential Loyalists
    (rfm_df["R_score"] >= 4) &
    (rfm_df["F_score"].between(2, 3)),

    # New Customers
    (rfm_df["R_score"] >= 4) &
    (rfm_df["F_score"] == 1),

    # At Risk
    (rfm_df["R_score"] <= 2) &
    (rfm_df["F_score"] >= 4),

    # Hibernating
    (rfm_df["R_score"] <= 2) &
    (rfm_df["F_score"].between(2, 3)),

    # Lost Customers
    (rfm_df["R_score"] <= 2) &
    (rfm_df["F_score"] == 1)

]
```

```python
choices = [
    "Champions",
    "Loyal Customers",
    "Potential Loyalists",
    "New Customers",
    "At Risk",
    "Hibernating",
    "Lost Customers"
]
```

```python
rfm_df["Segment"] = np.select(
    conditions,
    choices,
    default="Regular Customers"
)
```

## Business insights

## Key Business Insights

- Champions represent the company's highest-value customers.
- Lost Customers have the highest recency and lowest purchase frequency.
- Potential Loyalists can be converted into Loyal Customers through targeted campaigns.
- Regular Customers form the largest stable customer base.
- At Risk customers require immediate retention strategies.
- New Customers should receive onboarding offers and personalized recommendations.
- Loyal Customers are ideal candidates for upselling and cross-selling.
- Hibernating Customers can be re-engaged through email marketing and discounts.

| Segment             | Business Strategy               |
| ------------------- | ------------------------------- |
| Champions           | VIP rewards, exclusive offers   |
| Loyal Customers     | Upsell & Cross-sell             |
| Potential Loyalists | Loyalty programs                |
| New Customers       | Welcome offers                  |
| Regular Customers   | Seasonal promotions             |
| At Risk             | Win-back campaigns              |
| Hibernating         | Re-engagement emails            |
| Lost Customers      | Low-cost reactivation campaigns |
