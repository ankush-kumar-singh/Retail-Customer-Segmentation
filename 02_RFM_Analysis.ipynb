## RFM SEGMENTATION

```python
# INdustry rfm segments
import numpy as np
```

```python
rfm_df[rfm_df["Segment"] == "Others"][
    ["Recency", "Frequency", "Monetary",
     "R_score", "F_score", "M_score", "Score"]
].head(30)
```

```python
rfm_df["R_score"] = rfm_df["R_score"].astype(int)
rfm_df["F_score"] = rfm_df["F_score"].astype(int)
rfm_df["M_score"] = rfm_df["M_score"].astype(int)
```

```python
conditions = [

    # Champions
    (rfm_df["R_score"] == 5) &
    (rfm_df["F_score"] >= 4) &
    (rfm_df["M_score"] >= 4),

    # Loyal Customers
    (rfm_df["R_score"] >= 4) &
    (rfm_df["F_score"] >= 4),

    # Potential Loyalists
    (rfm_df["R_score"] >= 4) &
    (rfm_df["F_score"].between(2, 3)),

    # New Customers
    (rfm_df["R_score"] >= 4) &
    (rfm_df["F_score"] == 1),

    # At Risk
    (rfm_df["R_score"] <= 2) &
    (rfm_df["F_score"] >= 4),

    # Hibernating
    (rfm_df["R_score"] <= 2) &
    (rfm_df["F_score"].between(2, 3)),

    # Lost Customers
    (rfm_df["R_score"] <= 2) &
    (rfm_df["F_score"] == 1)

]
```

```python
choices = [
    "Champions",
    "Loyal Customers",
    "Potential Loyalists",
    "New Customers",
    "At Risk",
    "Hibernating",
    "Lost Customers"
]
```

```python
rfm_df["Segment"] = np.select(
    conditions,
    choices,
    default="Regular Customers"
)
```
